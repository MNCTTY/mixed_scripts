```python
import os
import glob
```

```python
path = 'Fin/'
```

```python
files = glob.glob(path+'*')
```

```python
files[:10]
```

```python
names = []
for file in files:
    names.append(file.split('/')[-1])
names[:10]
```

```python
to_del = []
for name in names:
    if len(name)>16:
        to_del.append(name)
```

```python
len(to_del)
```

```python
for name in to_del:
    os.remove(path+name)
```

```python
new_files = glob.glob(path+'*')
```

```python
new_files[:10]
```

```python
new_names = []
for file in new_files:
    new_names.append(file.split('/')[-1])
new_names[:10]
```

```python
to_corr = []
good = []
for name in new_names:
    if len(name)>12:
        to_corr.append(name)
    else:
        good.append(name)
```

```python
to_corr[:1]
```

```python
good[:1]
```

```python
for name in to_corr:
    corrected = name[:-7]+name[-4:]
    os.rename(path+name, path+corrected)
```
